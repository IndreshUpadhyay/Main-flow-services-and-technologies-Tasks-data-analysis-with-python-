```python
import pandas as pd
import matplotlib.pyplot as plt
d1=pd.read_excel("/content/sample_data/tips.csv")

# scatter plot with day aganist
plt.scatter(d1['day'],d1['tip'])

# adding title to plot
plt.title("scatter plot")
 # seting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
d1=pd.read_excel("/content/sample_data/tips.csv")

# scatter plot with day aganist
plt.scatter(d1['day'],d1['tip'],s=d1['total_bill'])
plt.title("scatter plot")
plt.xlabel('Day')
plt.ylabel('Tip')
plt.colorbar()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
d1=pd.read_excel("/content/sample_data/tips.csv")
plt.plot(d1['tip'])
plt.plot(d1['size'])
plt.title("Scatter plot")
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
d1=pd.read_excel("/content/sample_data/tips.csv")
plt.bar(d1['day'],d1['tip'])
plt.title("Bar Chart")
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
d1=pd.read_excel("/content/sample_data/tips.csv")
plt.hist(d1['size'])
plt.hist(d1['tip'])
plt.title("Histrogram Chart")
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```
